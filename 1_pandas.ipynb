**Задача 1**\
Сгенерировать массив из случайных положительных чисел произвольной длины. Вывести элементы, которые больше среднего арифметического.

```python
import random
np.random.seed(123)
arr = np.random.rand(10)
new_arr = []
for i in range(len(arr)):
    if arr[i-1] > arr.mean():
        new_arr.append(float(arr[i-1]))
print('Сгенерированный массив:', arr,'\nСреднее арифмеическое:', arr.mean(),'\nЭлементы, которые больше среднего арифметического:',new_arr)
```

**Задача 2**\
Сгенерировать массив из 20 случайных элементов, элементы которого лежат в интервале [0, 10].
+ Определить количество элементов, отличающихся от минимального на 5.
+ Найти среднее арифметическое всех четных элементов массива, стоящих на нечетных местах.

```python
# 1.1
np.random.seed(23)
arr_task2_size = 20
arr_task2 = [np.random.randint(0, 11) for _ in range(arr_task2_size)]
arr_task2
```

```python
# 1.2
diff_num_counter = 0
for _ in range(len(arr_task2)):
    if arr_task2[_] == np.min(arr_task2) + 5:
        diff_num_counter += 1
print('В сгенерированном массиве', diff_num_counter, 'элементов, отличающихся от минимального на 5.')
```

```python
# 1.3
arr_task2_odds = []
for i in range(len(arr_task2[0::2])):
    if arr_task2[0::2][i] % 2 == 0:
        arr_task2_odds.append(arr_task2[0::2][i])

print('Среднее арифметическое всех четных элементов массива, стоящих на нечетных местах:', float(np.mean(arr_task2_odds)))
```

**Задача 3**\
Сгенерировать матрицу A размером 5 на 5, элементы которой лежат в интервале [1, 10].
+ Возвести в квадрат максимальный элемент матрицы.
+ Все четные числа заменить на 0.

```python
# 1.1
arr_task3_size = 25
arr_task3 = [np.random.randint(0, 11) for _ in range(arr_task3_size)]
arr_task3_matrix = np.array(arr_task3).reshape(5, 5)
arr_task3_matrix
```

```python
# 1.2
max_x = max(arr_task3_matrix, key=max)
max_arr = float(np.max(max_x))
arr_task3_matrix[np.where(arr_task3_matrix == max_arr)] = arr_task3_matrix[np.where(arr_task3_matrix == max_arr)] **2
arr_task3_matrix
```

```python
# 1.3
i, j = 0, 0
for i in range(5):
    for j in range(5):
        if arr_task3_matrix[j][i] % 2 == 0:
            arr_task3_matrix[j][i] = 0
arr_task3_matrix
```

**Задача 4**\
Сгенерировать матрицу B размером 5 на 5, элементы которой лежат в интервале [-5, 5]. Отрицательные элементы матрицы заменить на число отрицательных элементов в строке, в которой они находятся.

```python
# 1.1
arr_task4_size = 25
arr_task4 = [np.random.randint(-5, 6) for _ in range(arr_task4_size)]
arr_task4_matrix = np.array(arr_task4).reshape(5, 5)
arr_task4_matrix
```

**Задание 2**
+ Выведите распределение переменной __pclass__ для мужчин и женщин.
    + Сколько мужчин было в первом классе?
    + Сколько женщин было в третьем классе?
      
+ Создайте новый признак __age_group__, который равен:
    + 1, если пассажир был моложе 30 лет;
    + 2, если пассажиру было от 30 до 60 лет;
    + 3, если пассажир был старше 60 лет.

```python
# 1.1
pclass_sex = df.pivot_table(index='pclass', columns='sex', values='name', aggfunc='count')
pclass_sex # распределение pclass для мужчин и женщин
```

```python
# 1.2
print(len(df[(df['sex']=='male') & (df['pclass']==1)]), 'мужчин было в первом классе','\n',len(df[(df['sex']=='female') & (df['pclass']==3)]), 'женщин было в третьем классе')
```

```python
# 2
def different_age(x):
    if x < 30:
        return 1 # моложе 30 лет
    elif x >= 60:
        return 3 # старше 60 лет
    else:
        return 2 # от 30 до 60 лет
df['age_group'] = df['age'].apply(different_age)

df.head()
```
